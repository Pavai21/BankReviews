```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get("https://www.bankbazaar.com/reviews/hdfc-bank/all-products.html")
```

```python
soup=BeautifulSoup(res.text,'lxml')
```

```python
bank_name = []
Name_html = soup.find_all("div",{"class":"review-bank-title"})
```

```python
for i in Name_html:
    bank_name.append(i.find("img").get("alt"))
```

```python
Customer_name = []
Cus_name = soup.find_all(class_='js-author-name')
```

```python
for i in Cus_name:
    Customer_name.append(i.text)
```

```python
Reviews = []
Review_html = soup.find_all(class_="text_here review-desc-more")
for i in Review_html:
    Reviews.append(i.text)
```

```python
Services=[]
Services_html=soup.find_all("a",{"class":"user-review-comment js-individual-title"})
```

```python
for i in Services_html:
    Services.append(i.text)
```

```python
dict1={'bank_name':bank_name,'Customer_name':Customer_name,'Reviews':Reviews,"Services":Services}
```

```python
df = pd.DataFrame.from_dict(dict1)
```
